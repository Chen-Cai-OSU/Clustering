```python
def get_stairs(points, distm, den=False):
    if den is False:
        f = - density(points).reshape(len(points),1)
    else:
        f =  density(points).reshape(len(points),1)
    
    stairs = []
    show_flag = False
    save_flag = False
    idx_set = range(len(f))

    for idx in idx_set:
        stair = get_stair(f, distm, idx=idx, method='heu')
        stairs.append(stair)
    return stairs
```
